```python
import pandas as pd
import matplotlib.pyplot as plt
import io

#1 leitura dos arquivos
csv_data = """ano,mes,despesa,valor_despesa
2024,janeiro,alimentação,240.00
2024,janeiro,transporte,130.00
2024,fevereiro,alimentação,250.00
2024,fevereiro,transporte,140.00
2024,março,alimentação,230.00
2024,março,transporte,120.00
2024,abril,alimentação,245.00
2024,abril,transporte,145.00
2024,maio,alimentação,260.00
2024,maio,transporte,150.00
2024,junho,alimentação,225.00
2024,junho,transporte,138.00"""

df = pd.read_csv(io.StringIO(csv_data))

#2 exibir as 5 primeiras linhas e os tipos de imformacoes
print("5 primeiras linhas do dataframe:")
print(df.head())
print("\ninformacaoes de tipos de dados:")
print(df.info())

#3 convertendo as colunas mes para o tipo categorico e ordenado

meses_ordenados = ['janeiro', 'fevereiro', 'março', 'abril', 'maio', 'junho']
df['mes'] = pd.Categorical(df['mes'], categories=meses_ordenados, ordered=True)

#4 calculo total das despesas agrupada por mes
gastos_por_mes = df.groupby('mes', observed=True)['valor_despesa'].sum()
gastos_por_categoria = df.groupby('despesa')['valor_despesa'].sum()

#grafico 1: despesas por mes
plt.figure(figsize=(10,  5))
gastos_por_mes.plot(kind='bar', color='skyblue')
plt.title('despesas por mes')
plt.xlabel('mes')
plt.ylabel('valor_despesa')
plt.xticks(rotation=45)
plt.show()


#grafico 2: despesas por categoria

plt.figure(figsize=(8, 5))
gastos_por_categoria.plot(kind='bar', color='orange')
plt.title('total de despesas por categoria')
plt.xlabel('categoria')
plt.ylabel('valor_despesa')
plt.xticks(rotation=0)
plt.show()
```

ANALISE DAS DESPESAS (JANEIRO - JUNHO)

GASTOS MENSAIS
com base nos dados, o mes de maio foi o que tave maior volume de despesas,os meses de feveriero e abril aparecem logo em seguida, o periodo de menor consumo foi marco.

a analise de categoria de alimentacao aprensentou o maior numero de despesas, a categoria de transporte teve um impacto menor no orcamento.
